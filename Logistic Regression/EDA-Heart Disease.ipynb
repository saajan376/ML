```python
#importing the necessary libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading the dataset
df = pd.read_csv('heart.csv')
df
```

```python
#to get the information about the data, their values and data types
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#since slope, thal and ca contain more than 50% of null values, we will drop them
df.drop(['slope', 'thal', 'ca'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
numerical_cols = ['trestbps', 'chol', 'thalch', 'oldpeak']
for col in numerical_cols:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)
df.isnull().sum()
```

```python
#we will be filling in the columns fbs, restecg and exang with mode values
categorical_cols = ['fbs', 'restecg', 'exang']
for col in categorical_cols:
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)

df.isnull().sum()
```

```python
corr_matrix = df.corr(numeric_only=True)
print(corr_matrix)
```

```python
#oldpeak and thalch are highly correlated with target variable
#chol and trestbps are also correlated with target variable but not as much as oldpeak
```
