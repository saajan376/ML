```python
#importing the necessary libraries


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading the dataset
df=pd.read_csv("BostonHousing.csv")

#to change the floating point values to something else and here we've changed it to 2 decimal places
#this is useful for better readability of the data
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df
```

```python
#to get the information about the data, their values and data types
df.info()
```

```python
#to get another form of description about the dataset
df.describe()
```

```python
#to find the missing values in the dataset
df.isnull().sum()
```

```python
#to find the unique values in the dataset
df.nunique()
```

```python
#since rn column has 5 missing values, I am going to add the mean value of the column to those 5 missing values
mean_rm = df['rm'].mean()
df['rm'].fillna(mean_rm, inplace=True)
```

```python
df.isnull().sum()  # to check if the missing values are filled
```

```python
#visualizing using histograms
df.hist(bins=20,figsize=(20, 15)) 
```

```python
# 1. Create a scatter plot for 'rm' vs 'medv'
plt.figure(figsize=(10, 6))
sns.regplot(x='rm', y='medv', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Rooms vs. Home Value', fontsize=16)
plt.xlabel('Average Number of Rooms (rm)')
plt.ylabel('Median Home Value in $1000s (medv)')
plt.savefig('rooms_vs_value_scatter.png')
print("Saved scatter plot to rooms_vs_value_scatter.png")
```

```python
# 2. Create a scatter plot for 'lstat' vs 'medv'
plt.figure(figsize=(10, 6)) # Create a new figure to avoid plotting on the same axes
sns.regplot(x='lstat', y='medv', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Population Status vs. Home Value', fontsize=16)
plt.xlabel('% Lower Status of Population (lstat)')
plt.ylabel('Median Home Value in $1000s (medv)')
plt.savefig('lstat_vs_value_scatter.png')
print("Saved scatter plot to lstat_vs_value_scatter.png")
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
```
